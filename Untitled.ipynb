```python
import csv
import requests
import csv
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('fivethirtyeight')

CSV_URL = 'https://query1.finance.yahoo.com/v7/finance/download/%5ENSEI?period1=1591415995&period2=1622951995&interval=1d&events=history&includeAdjustedClose=true'

with requests.Session() as s:
    download = s.get(CSV_URL)
    decoded_content = download.content.decode('utf-8')
    cr = csv.reader(decoded_content.splitlines(), delimiter=',')
    my_list = list(cr)
    my_list[0].append('returns')
    n=len(my_list)
    returns=[]
    mn=0
    mx=0 
    for i in range(1,n):
        open_price=my_list[i][1]
        close_price=my_list[i][4]
        if open_price=='null' or close_price=='null':
            ret=0
        else:
            ret=float(close_price)-float(open_price)
            ret=ret/float(open_price)
        if ret<mn:
            mn=int(ret)
        if ret<mx:
            mx=int(ret)
        my_list[i].append(round(ret,4))
        returns.append(round(ret,4))
    for row in my_list:
        print(row) 
    bins=[]
    for i in range(-1,1):
        nxt1=i+1
        while(i<nxt1):
            i+=0.01
            bins.append(i)
    plt.hist(returns,bins=bins,edgecolor='black')
    plt.title('NSE historical data analysis using histogram')
    plt.xlabel('Returns Distribution')
    plt.ylabel('Frequency')
    plt.tight_layout()
    
```
